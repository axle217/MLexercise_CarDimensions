```python
# Setup directories for Google Colab
import os
os.chdir('/content/drive/MyDrive/Colab Notebooks')
%cd MLexercise_CarDimensions

import glob
import re
import pandas as pd

# pd.set_option('display.max_rows', None)
# pd.set_option('display.max_columns', None)

file_paths = glob.glob("data/2*_en.csv")
file_paths.sort(key=lambda x: int(re.search(r'\d+', x).group()))
print(file_paths)

dataframes = [pd.read_csv(path, encoding='ISO-8859-1') for path in file_paths]
```

```python
removed_NaN_dataframes = [df.dropna() for df in dataframes]

merged_data = pd.concat(removed_NaN_dataframes)
# Remove non-breaking white spaces in string type entries
merged_data = merged_data.map(
    lambda x: x.strip().replace('\xa0', '') if isinstance(x, str) else x
)
```

```python
# Check for white spaces in label
print(merged_data.columns.tolist())
merged_data.columns = (
    merged_data.columns
    .str.strip()  # Remove leading/trailing whitespace
    .str.replace('\xa0', '', regex=False)  # Remove non-breaking spaces
    # .str.replace(' ', '_')  # Replace spaces with underscores
)
print(merged_data.columns.tolist())

merged_data = merged_data.rename(columns={"MYR": "Model Year",
                            "OL": "Overall Length",
                            "OW": "Overall Width",
                            "OH": "Overall Height",
                            "WB": "Wheelbase",
                            "CW": "Curb Weight",
                            "A": "Front End Length",
                            "B": "Rear End Length",
                            "C": "Side Glass Height",
                            "D": "Body Side Height",
                            "E": "Roof Width",
                            "F": "Front Overhang",
                            "G": "Rear Overhang",
                            "TWF": "Track Width Front",
                            "TWR": "Track Width Rear",
                            "WDIST": "Weight Distribution"
})

print(merged_data.columns.tolist())
```

```python
# print(merged_data['MAKE'].unique())
# print(merged_data['MODEL'].unique())
print(merged_data['Wheelbase'].unique())
```

```python
# Convert mixed string to numeric
merged_data['Wheelbase'] = pd.to_numeric(merged_data['Wheelbase'], errors='coerce')
print(merged_data['Wheelbase'].unique())

print(merged_data.dtypes)
```

```python
# Cleaning typos and duplicates
merged_data.loc[:,'MAKE'] = merged_data['MAKE'].str.upper().str.replace('\xa0', '').str.replace(' ', '')
merged_data.loc[:,'MODEL'] = (
    merged_data['MODEL'].str.upper()
      .str.replace(' ', '')
      .str.replace('-', '')
      .str.replace('.', '')
      .str.replace('/', '')
)

merged_data = merged_data.drop_duplicates()
```

```python
merged_data = merged_data.sort_values(by=['MODEL'])
merged_data
# print(sorted(merged_data['MAKE'].unique()))
```

```python
# merged_data['MAKE'].value_counts()
```

```python
merged_data[['Front Weight %', 'Rear Weight %']] = (
    merged_data['Weight Distribution'].str.split('/', expand=True).astype(float)
)
```

```python
grouped_data = (
    merged_data
    .groupby('MODEL', as_index=False)
    .agg({
        "MAKE" : "first",
        "Model Year" : "max", # There are same models with different years
        "Overall Length" : "max",
        "Overall Width" : "max",
        "Overall Height" : "max",
        "Wheelbase" : "max",
        "Curb Weight" : "max",
        "Front End Length" : "max",
        "Rear End Length" : "max",
        "Side Glass Height" : "max",
        "Body Side Height" : "max",
        "Roof Width" : "max",
        "Front Overhang" : "max",
        "Rear Overhang" : "max",
        "Track Width Front" : "max",
        "Track Width Rear" : "max",
        "Front Weight %" : "max",
        "Rear Weight %" : "max"
    })
)

grouped_data = grouped_data.sort_values(by=['MODEL'])
grouped_data
```
