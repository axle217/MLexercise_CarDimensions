```python
# Setup directories for Google Colab
import os
os.chdir('/content/drive/MyDrive/Colab Notebooks')
%cd MLexercise_CarDimensions

import glob
import re
import pandas as pd

# pd.set_option('display.max_rows', None)
# pd.set_option('display.max_columns', None)

file_paths = glob.glob("data/2*_en.csv")
file_paths.sort(key=lambda x: int(re.search(r'\d+', x).group()))
print(file_paths)

dataframes = [pd.read_csv(path, encoding='ISO-8859-1') for path in file_paths]
```

```python
# Data cleaning
# cleaned_dataframes = dataframes

cleaned_dataframes = [df.dropna() for df in dataframes]
for i in range(len(cleaned_dataframes)):
  cleaned_dataframes[i] = cleaned_dataframes[i].drop_duplicates()
# cleaned_dataframes[0]

merged_data = pd.concat(cleaned_dataframes)
# merged_data
```

```python
# Check for white spaces in label
print(merged_data.columns.tolist())
merged_data.columns = (
    merged_data.columns
    .str.strip()  # Remove leading/trailing whitespace
    .str.replace('\xa0', '', regex=False)  # Remove non-breaking spaces
    # .str.replace(' ', '_')  # Replace spaces with underscores
)
print(merged_data.columns.tolist())

merged_data = merged_data.rename(columns={"MYR": "Model Year",
                            "OL": "Overall Length",
                            "OW": "Overall Width",
                            "OH": "Overall Height",
                            "WB": "Wheelbase",
                            "CW": "Curb Weight",
                            "A": "Front End Length",
                            "B": "Rear End Length",
                            "C": "Side Glass Height",
                            "D": "Body Side Height",
                            "E": "Roof Width",
                            "F": "Front Overhang",
                            "G": "Rear Overhang",
                            "TWF": "Track Width Front",
                            "TWR": "Track Width Rear",
                            "WDIST": "Track Width"
})

print(merged_data.columns.tolist())
```
